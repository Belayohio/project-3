```python
#importing the dependency
import pandas as pd
import numpy as np
# copy -on-write mode
pd.options.mode.copy_on_write = True
```

```python
# assign variable for russian file
russian_csv="russia_losses.csv"
```

```python
#read the csv
file_df=pd.read_csv(russian_csv,header=0)
file_df.head()
```

```python
#check the file info
file_df.info()
```

```python
# droping "unit" column.
cleaned_df=file_df.drop(["unit"],axis=1)


#drop null value 
df = cleaned_df.dropna()
df.head()
```

```python
#check the df info after dropped the Null value
df.info()
```

```python
#explore the columns
df.columns
```

```python
# loop through "tags" column and extract using comma delimiter 

# list comprehension 
tag= [i for i in df["tags"].str.split(',').tolist()]
        
print(len(tag))
print('---------------------------------------------')
print(tag[0:50])
```

```python
# extract the  id columns.
id = [d for d in df['id']]
```

```python
print(len(id))
```

```python
# convert the "tag" list to dataframe
tag_df = pd.DataFrame(tag)
tag_df.head()
```

```python
# renaming the column to "tag0"-"tag5" and its ID 
tag_df=pd.DataFrame({"id":id,'tag0':tag_df[0],'tag1':tag_df[1],'tag2':tag_df[2],'tag3':tag_df[3],'tag4':tag_df[4],'tag5':tag_df[5]})
```

```python
#display tag_df

tag_df.head()
```

```python
# filter the tag0 and its Id.
df_df0=tag_df[["id",'tag0']].loc[tag_df['tag0']>=""]
df_df0.head()
```

```python
# filter the tag1 and its Id.
df_df1=tag_df[["id",'tag1']].loc[tag_df['tag1']>=""]
df_df1.head()
```

```python
# filter the tag2 and its Id.
df_df2=tag_df[["id",'tag2']].loc[tag_df['tag2']>=""]
df_df2.head()
```

```python
# filter the tag3 and its Id.
df_df3=tag_df[["id",'tag3']].loc[tag_df['tag3']>=""]
df_df3.head()
```

```python
# filter the tag4 and its Id.
df_df4=tag_df[["id",'tag4']].loc[tag_df['tag4']>=""]
df_df4.head()
```

```python
# filter the tag5 and its Id.
df_df5=tag_df[["id",'tag5']].loc[tag_df['tag5']>=""]
df_df5.head()
```

```python
# Save the dataframe to csv files

df_df0.to_csv("tag0.csv",index=False)
df_df1.to_csv("tag1.csv",index=False)
df_df2.to_csv("tag2.csv",index=False)
df_df3.to_csv("tag3.csv",index=False)
df_df4.to_csv("tag4.csv",index=False)
df_df5.to_csv("tag5.csv",index=False)
```
